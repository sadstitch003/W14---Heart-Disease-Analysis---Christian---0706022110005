```python
import pandas as pd

df = pd.read_csv('heart.csv')
df.head()
```

```python
df.describe()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df))
df_scaled.columns = df.columns
df_scaled.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 4))
sns.histplot(data=df, x='age', bins=20)
plt.title("Age Distribution")
plt.show()
```

> Chart diatas menunjukkan bahwa usia pada dataset terdistribusi secara normal 

```python
plt.figure(figsize=(8,8))
sns.heatmap(data=df.corr(), vmin=-1, vmax=1, cmap='coolwarm', annot=True, fmt='.2f')
plt.title("Heart Dataset Corelation")
plt.show()
```

> Attribut paling berpengaruh terhadap output adalah cp (Chest pain type), thalachh (Maximum heart rate achieved), exng (Exercise induced angina 1 = yes 0 = no), dan oldpeak

```python
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
sns.countplot(data=df, x='cp')
plt.xticks(ticks=[0, 1, 2, 3], labels=['Typical Angina', 'Atypical Angina', 'Non-anginal Pain', 'Asymptomatic'], rotation=45)
plt.title('Chess Pain Type Distribution')
plt.xlabel('')

plt.subplot(1, 2, 2)
sns.countplot(data=df, x='exng')
plt.title('Maximum Heart Rate Achieved Distribution')
plt.xticks(ticks=[0, 1], labels=['No','Yes'])
plt.xlabel('Maximum Heart Rate Achieved')
plt.show()
```

```python
sns.countplot(data=df, x='sex', palette=['blue', 'blue'])
sns.countplot(data=df[df['output'] == 1], x='sex', palette=['orange', 'orange'])
plt.xticks(ticks=[0, 1], labels=['Male', 'Female'])
plt.title('Distribution of Male vs Female')
plt.legend(labels=['Less chance of heart attack', 'More chance of heart attack'])
plt.show()
```

> Persentase laki laki yang memiliki peluang mendapat serangan jantung lebih besar dari pada wanita

```python
sns.scatterplot(data=df, x="age", y="chol")
plt.title("Corelation of Age and Cholestrol")
plt.show()
```

> Chart diatas menunjukkan bahwa tidak ada korelasi antara kolestrol dengan usia. Usia muda bisa saja memiliki kolestrol tinggi dan begitu pula sebaliknya

```python
from sklearn.model_selection import train_test_split

x = df_scaled.drop(columns=['output'])
y = df['output']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix

model = LogisticRegression(random_state=0)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)

logistic_accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print(f"Accuracy: {logistic_accuracy}")
print("Confusion Matrix:")
print(conf_matrix)
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(random_state=0)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)

decisiontree_accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

# Display the results
print(f"Accuracy: {decisiontree_accuracy}")
print("Confusion Matrix:")
print(conf_matrix)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(random_state=0)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)

randomforest_accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print(f"Accuracy: {randomforest_accuracy}")
print("Confusion Matrix:")
print(conf_matrix)
```

```python
import xgboost as xgb

model = xgb.XGBClassifier(random_state=0)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)

xgb_accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print(f"Accuracy: {xgb_accuracy}")
print("Confusion Matrix:")
print(conf_matrix)
```

```python
accuracy_scores = [logistic_accuracy, decisiontree_accuracy, randomforest_accuracy, xgb_accuracy]
model_names = ['Logistic Regression', 'Decision Tree', 'Random Forest', 'XGBoost']

sns.barplot(x=accuracy_scores, y=model_names)
plt.xlabel('Accuracy')
plt.ylabel('Model')
plt.title('Accuracy Scores for Different Models')
plt.xlim(0, 1)
plt.show()
```
